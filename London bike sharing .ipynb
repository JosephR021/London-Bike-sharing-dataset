```python
#creating the dictionary to rename the column names

bikes_dict ={
    'timestamp' : 'time',
    'cnt' : 'count',
    't1' : 'temperature_real',
    't2' : 'temperature_feel',
    'hum': 'humidity',
    'wind_speed' : 'wind_speed_kmp',
    'weather_code' : 'weather',
    'is_holiday' : 'is_holiday' ,
    'is_weekend' : 'is_weekend',
    'season': 'season'
    }
#renaming the column names by replacing the original name by dictionary created

bikes.rename(bikes_dict , axis=1, inplace=True)
```

```python
#changing the humidity value to percentage by dividing the values by 100 ie.(value/100)

bikes.humidity =  bikes.humidity / 100 
```

```python
#creating the dictionary to give the actual season name instead of integer

season_dict = {
    '0.0' : 'spring',
    '1.0' : 'summer',
    '2.0' : 'autumn',
    '3.0' : 'winter',
}

# creating the dictionary to give the actual weather instead of integer

weather_dict= {
    '1.0': 'clear',
    '2.0' : 'scattered cloud',
    '3.0' : 'broken clouds',
    '4.0' : 'cloudy',
    '7.0' : 'rain',
    '10.0' : 'rain with thunderstrom',
    '26.0': 'snowfall'
}
```

```python
#to know the data type of each column

bikes.dtypes
```

```python
#changing the season data type from float to str

bikes.season = bikes.season.astype('str')

#changing the weather data type from float to str

bikes.weather = bikes.weather.astype('str')
```

```python
#mapping the season_dict to season column

bikes.season = bikes.season.map(season_dict)

#mapping the season_dict to weather column

bikes.weather = bikes.weather.map(weather_dict)
```

```python
#cheaking whether the mapping values are updated to the dataframe

bikes.head(20)
```

```python
#filtering the season column if needed

filter_spring =bikes[bikes['season'] == 'spring']

filter_winter =bikes[bikes['season'] == 'winter']

filter_summer =bikes[bikes['season'] == 'summer']

filter_autumn =bikes[bikes['season'] == 'autumn']
```

```python
#Converting the time column from str to datetime

bikes.time = bikes.time.astype('datetime64[ns]')
```

```python
bikes.dtypes
```

```python
#Extrating the month name from the time column by adding a new column

bikes['Month'] = bikes.time.dt.strftime('%B')

#Extrating the day name from the time column by adding a new column

bikes['Day'] = bikes.time.dt.strftime('%A')

#Extrating the year from the time column by adding a new column

bikes['Year'] = bikes.time.dt.year
```

```python
#Writing the dataframe to excel to visuilize the data using power BI

bikes.to_excel('london_bikes_updated1.xlsx', sheet_name='london_bikes_data', index=False)
```
